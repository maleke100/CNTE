```python
!pip install gradio openai-whisper
import gradio as gr

def repondre(audio):
    # transcription automatique de Gradio avec Whisper
    import whisper
    import tempfile
    import os

    # sauvegarde temporaire du fichier audio
    temp = tempfile.NamedTemporaryFile(delete=False, suffix=".wav")
    # Assuming audio is a file-like object or a file path
    if isinstance(audio, str): # If audio is a file path
        with open(audio, 'rb') as audio_file:
            temp.write(audio_file.read())
    else: # Assuming audio is a file-like object
        temp.write(audio.read())

    temp.close()

    # transcription avec Whisper (modèle tiny pour rapidité)
    model = whisper.load_model("tiny")
    result = model.transcribe(temp.name, language="fr")
    texte = result["text"].lower()

    os.remove(temp.name)  # nettoyage

    # logique de réponse
    if "bonjour" in texte:
        reponse = "Bonjour à toi !"
    elif "triste" in texte:
        reponse = "Courage, ça ira mieux !"
    elif "heureux" in texte:
        reponse = "Génial, je suis content pour toi !"
    elif "fâché" in texte or "faché" in texte:
        reponse = "Respire un bon coup, ça va aller."
    else:
        reponse = "Merci pour ton message."

    return reponse

gr.Interface(
    fn=repondre,
    inputs=gr.Audio(type="filepath", label="Parle ici"),
    outputs="text",
    title="Assistant Vocal Simple",
    description="Dis bonjour ou ton humeur.",
).launch()
```
